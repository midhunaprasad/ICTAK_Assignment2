```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
sns.set()
filename='diabetes_model.csv'
df=pd.read_csv(filename)
pd.set_option('display.max_rows', None)
df
```

# 1

```python
plt.figure(figsize=(8,6))
plt.title('Count of Outcome')
ax=sns.countplot(x=df['Outcome'])
plt.xlabel('Outcomes')
plt.ylabel('Frequency')
for p in ax.patches:
        ax.annotate('%{:.1f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()+5))
```

# 2

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Outcome'],y=df['Age'],data=df)
```

As you can clearly see from the above boxplot the median value of outcome lies below that of age and we can see a lot of outlayers, so we  can say as age increases the chance to get diabetes increases.

```python
plt.figure(figsize=(8,6))
plt.title('Heat Map')
sns.heatmap(df.corr(),annot=True,cmap="Blues")
plt.plot()
```

```python
sns.boxplot(x=df['Outcome'],y=df['DiabetesPedigreeFunction'],data=df)
```
